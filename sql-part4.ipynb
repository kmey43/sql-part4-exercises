1. Write a query that will return the number of users who rated a book above it's average rating.

```sql
with avg_ratings as (
SELECT
r.book_id,
avg(r.rating) as avg_rating
from booksdb.dbo.ratings r
group by book_id
)

SELECT
count(user_id) as num_users
from booksdb.dbo.ratings r
join avg_ratings ar on ar.book_id=r.book_id
where r.rating > ar.avg_rating
--group by r.book_id
--having r.rating > ar.avg_rating

```

2. Write a query that returns the book ids of all books that have over 1000 ratings of 1 star or over 1000 ratings of 5 stars.

```sql
with one_star as (
SELECT
r.book_id,
count(rating) as one_count_ratings
from BooksDB.dbo.ratings r
where rating=1 or rating=5
group by book_id
having count(rating)>80
),

five_star as (
SELECT
r.book_id,
count(rating) as five_count_ratings
from BooksDB.dbo.ratings r
where rating=5
group by book_id
having count(rating)>80
--order by five_count_ratings desc
)

select 
fs.book_id,
os.one_count_ratings,
fs.five_count_ratings
from five_star fs 
inner  join one_star os on fs.book_id=os.book_id
order by fs.five_count_ratings desc


```

4. Write a query that returns the book ids of books that have a language code of "en-US" and not a langugae code of "en-GB".

```sql
with en_us as (
select*
from booksdb.dbo.books
where language_code='en-US')

select
book_id,
language_code
from en_us
where language_code <> 'en-GB'
```

5. Write a query that returns the names of the tags and the tag ids for tags that were used over 100,000 times for a book. 

```sql
 with tag_counts as (
select
bt.tag_id,
tg.tag_name,
count(b.book_id) as tag_count
from BooksDB.dbo.books b
join booksdb.dbo.book_tags bt on bt.goodreads_book_id=b.book_id
join booksdb.dbo.tags tg on tg.tag_id=bt.tag_id
group by bt.tag_id, tg.tag_name
--order by tag_count desc
 )

select
tag_id,
tag_name,
tag_count
from tag_counts 
where tag_count>50
order by tag_count desc
```
